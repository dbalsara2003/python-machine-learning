# K-Means Clustering

## Importing required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
wine_data = pd.read_csv('winequality-red.csv')
X = wine_data.iloc[:, :-1].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Train the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters=6, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

## Visualize the clusters and their centroids

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(X[y_kmeans == 5, 0], X[y_kmeans == 5, 1], s = 100, c = 'yellow', label = 'Cluster 6')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Clusters of Wines')
plt.xlabel('Acidity')
plt.ylabel('Alcohol')
plt.legend()
plt.show()
```

## Create separate random test data for a new training

```python
X1 = np.random.uniform(5, 15, 300)
X2 = np.random.uniform(0.15, 1.5, 300)

X_test = np.column_stack((X1, X2))

# Plot the new random data in a scatter plot
plt.scatter(X_test[:, 0], X_test[:, 1], s = 100, c = 'red', label = 'Test Data')
plt.title('New Test Data')
plt.xlabel('Acidity')
plt.ylabel('Alcohol')
plt.legend()
plt.show()
```

## Add extra points in different regions of the data set to artificially create clusters in the dataset

```python
extraX1 = np.random.uniform(6, 7, 50)
extraY1 = np.random.uniform(0.2, 0.5, 50)
extraX2 = np.random.uniform(8, 9, 50)
extraY2 = np.random.uniform(0.2, 0.5, 50)
extraX3 = np.random.uniform(10, 11, 50)
extraY3 = np.random.uniform(0.2, 0.5, 50)
extraX4 = np.random.uniform(12, 13, 50)
extraY4 = np.random.uniform(0.2, 0.5, 50)
extraX5 = np.random.uniform(6.5, 7.5, 50)
extraY5 = np.random.uniform(0.6, 0.85, 50)
extraX6 = np.random.uniform(8, 9, 50)
extraY6 = np.random.uniform(0.6, 0.85, 50)

#Concatenate the new data to the original test data
X_test = np.concatenate((X_test, np.column_stack((extraX1, extraY1))), axis=0)
X_test = np.concatenate((X_test, np.column_stack((extraX2, extraY2))), axis=0)
X_test = np.concatenate((X_test, np.column_stack((extraX3, extraY3))), axis=0)
X_test = np.concatenate((X_test, np.column_stack((extraX4, extraY4))), axis=0)
X_test = np.concatenate((X_test, np.column_stack((extraX5, extraY5))), axis=0)
X_test = np.concatenate((X_test, np.column_stack((extraX6, extraY6))), axis=0)

plt.scatter(X_test[:, 0], X_test[:, 1], s = 100, c = 'red', label = 'Test Data')
plt.title('New Test Data')
plt.xlabel('Acidity')
plt.ylabel('Alcohol')
plt.legend()
plt.show()
```

## Train the new K-Means model on the new data and visualize the clusters in different colors

```python
kmeans_new = KMeans(n_clusters=6, init='k-means++', random_state=42)
kmeans_new.fit(X_test)
labels = kmeans_new.predict(X_test)

X_test1 = X_test[labels == 0]
X_test2 = X_test[labels == 1]
X_test3 = X_test[labels == 2]
X_test4 = X_test[labels == 3]
X_test5 = X_test[labels == 4]
X_test6 = X_test[labels == 5]

plt.scatter(X_test1[:, 0], X_test1[:, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_test2[:, 0], X_test2[:, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_test3[:, 0], X_test3[:, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X_test4[:, 0], X_test4[:, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X_test5[:, 0], X_test5[:, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(X_test6[:, 0], X_test6[:, 1], s = 100, c = 'yellow', label = 'Cluster 6')
plt.title('New Test Data')
plt.xlabel('Acidity')
plt.ylabel('Alcohol')
plt.legend()
plt.show()
```

## Use the clusters to predict which type of wines have the highest quality

```python
wine_quality = wine_data.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, wine_quality, test_size = 0.3, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

#Predict the test set results

y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
